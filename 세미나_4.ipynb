# 4. 선형 대수학 Linear Algebra

## 벡터 Vector

- 크기와 방향을 가지는 양을 나타내는 개념
- 일반적으로 화살표로 표시되며, 길이는 벡터의 크기를 나타내고, 방향은 벡터의 방향을 나타냄

1. 좌표 표현: n차원 공간에서의 벡터는 좌표로 표현될 수 있다.
2차원에서의 벡터 (x, y), 3차원에서의 벡터(x, y, z)
2. 컨포넌트 표현: 벡터는 각각의 차원에 대한 값을 가지는 컴포넌트로 표현될 수 있음.
2차원에서의 벡터 (x, y)는 각각 x, y 값은 벡터의 좌표 축에 대한 성분

- 벡터의 덧셈과 뺄셈

: 두 벡터의 각 성분을 독립적으로 더하거나 뺌

- 스칼라와의 곱셈

: 벡터에 스칼라(실수)를 곱하는 것.

벡터의 각 성분에 곱함(스칼라의 값이 음수이면 벡터의 방향이 반대로 될 수 있음)

- 내적(점곱)

: 두 벡터의 내적은 벡터 간의 곱셈 연산
.
내적은 각 성분의 곱을 합산한 것.

벡터의 길이와 벡터 간의 각도에 대한 정보를 제공

- 벡터의 크기

: 벡터의 길이를 나타냄.

각 성분의 제곱을 합산하고 그 후 제곱근을 취함.

벡터의 크기는 양수, 벡터의 방향과는 독립적

### 벡터의 내적

- 내적 결과가 0인 경우 두 벡터가 수직(직교)하다는 것을 의미(A · B = 0 -> 직교)

- 수직 성분 분해

두 벡터의 내적을 사용하면 한 벡터를 다른 벡터에 대한 수직 성분과 평행 성분으로 분해할 수 있음.

<내적을 이용하여 수직 성분을 구하는 법>

: 벡터 A의 수직 성분은 A - projᵦ(A)

projᵦ(A)는 벡터 A의 B에 대한 투영(projection)

projᵦ(A) = ((A · B) / (||B||²)) * B

- 코사인 유사도

: 두 벡터 사이의 각도에 대한 정보 제공

내적의 결과는 두 벡터의 길이와 두 벡터 사이의 각도의 코사인 값을 곱한 것과 동일함

A · B = ||A|| ||B|| cos(θ)

||A||: A의 길이, ||B||: B의 길이, θ: A와 B 사이 각도

내적은 다양한 분야에서 유용하게 활용

1. 벡터의 유사성 측정
2. 벡터의 정규화(normalization)
3. 벡터의 투영과 사영을 구하는 데 활용
4. 벡터 공간의 기저(basis)를 구성, 선형 종속성을 분석하는 데 중요한 도구임.


## 행렬 Matrix

: 행(row)와 열(column)로 이루어진 2차원 배열

m x n 행렬: m개의 행과 n개의 열을 가짐 크기를 나타내는 행렬의 차원은 m x n으로 표기

각각의 요소는 주로 대문자로 표현

Aij는 행렬 A의 i번째 행과 j번째 열에 위치한 요소를 나타냄

**행렬의 예시**

A = [1 2 3]
    [4 5 6]

2개의 행과 3개의 열 2 x 3 행렬 첫 번째 행의 요소는 1, 2, 3 두 번째 행의 요소는 4, 5, 6

- 행렬의 덧셈과 뺄셈

조건: 동일한 크기를 가진 두 행렬
    
A와 B가 동일한 크기를 가지는 행렬일 경우 A + B와 A - B는 각 요소별로 더하거나 뺀 결과를 가짐

(동일한 크기가 아닐 경우 정의되지 않음)

- 스칼라와의 곱셈

: 행렬의 모든 요소에 스칼라(실수) 값을 곱할 수 있음. 

행렬 A에 스칼라 c를 곱한 결과는 cA라 표기(각 요소에 c를 곱한 값을 가지는 행렬)

- 행렬의 곱셈

: 두 행렬의 크기와 규칙을 따름.

행렬 A의 열 수와 행렬 B의 행 수가 일치해야 가능

A와 B의 곱은 새로운 행렬 C 생성

행렬의 곱셈은 교환법칙이 성립하지 않음(AB ≠ BA)

- 항등 행렬

: 주 대각선 요소가 1이고, 나머지 요소는 모두 0인 정사각형 행렬

항등 행렬은 I로 표됨, 곱셈에서의 항등원과 유사한 역할을 함.

어떤 행렬 A와 항등 행렬 I를 곱하면 A는 변하지 않고 그대로 유지.

I = [1 0 0]
    [0 1 0]
    [0 0 1]

AI = IA = A

항등원(identity element): 임의의 원소(실수, 다항식, 행렬, 벡터 등)에 특정 연산을 했을 때 재귀시키는 연소

연산 결과 항등원이 나오게 하는 원소는 역원(inverse element)

- 전치 행렬

: 행과 열을 바꾼 새로운 행렬을 생성

주어진 행렬 A의 전치 행렬은 A^T로 표기

A = [1 2 3]

A^T = [1]
      [2]
      [3]

- 역행렬

: 정사각 행렬 A에 대해, 역행렬 A^(-1)은 A와 곱하면 항등 행렬 I를 얻을 수 있음.

역행렬은 존재하지 않을 수도 있으며, 존재할 경우 유일하지 않을 수 있음

행렬 A의 역행렬이 존재한다면 det(A) ≠ 0, 존재하지 않는다면 det(A) = 0

선형 방정식의 해를 구하는 등 다양한 응용에서 사용할 수 있음.

정사각행렬 A(n x n)

AB = In         BA = In        AB = BA = In

B: A의 역행렬 A^(=1)
 

> 선형 시스템, 선행 변환, 고유값과 고유 벡터 등 다양한 선형 대수학과 연관되어 있음
데이터 분석, 이미지 처리, 머신러닝, 컴퓨터 그래픽스 등 다양한 분야에 사용됨

* **역행렬을 사용하여 선형 방정식의 해를 구하는 방법 예**


**선형 방정식 2x + 3y = 8, 4x + 5y = 13**

위의 선형 방정식을 행렬과 벡터로 나타내면

[A][x] = [B]

여기서 A = [2 3] x = [x] b = [8]
           [4 5]     [y]     [13]
       
<br>    
       
**역행렬을 사용하여 x와 y 값 구하기**

우선 역행렬을 사용하기 위해선 행렬 A가 역행렬을 가지는지 확인해야 함.

역행렬이 존재하려면 행렬 A는 정방 행렬(square matrix) 즉, 행과 열의 개수가 동일해야 함

A 행렬의 행과 열의 개수는 2로 동일하므로 정방 행렬


<br> 

**역행렬을 구하기 위해 행렬 A의 determinant(행렬식) 계산하기**

det(A) = (2)(5) - (3)(4) = 10 - 12 = -2

det(A)가 0이 아닌 -2이기에 A는 역행렬을 가짐


<br>  

**역행렬 식: A^(-1) = (1/det(A)) * adj(A)**

adj(A): 여인수 행렬

adj(A) = [5 -3]
         [-4 2]
         
<br> 

**A 행렬의 역행렬 A^(-1) 구하기**
         
A^(-1) = (1/-2) * [5 -3]
                  [-4 2]


A^(-1) = [-5/2 3/2]
         [2 -1]
         
<br> 

**역행렬의 사용하여 x와 y의 값 구하기**

[x] = A^(-1) * [b]

[A^(-1)] = [-5/2 3/2]
           [2 -1]
           
[x] = [-5/2 3/2] * [8]
[y]   [2 -1]       [13]

계산하면

x = (-5/2)(8) + (3/2)(13) = -20 + 19.5 = -0.5
y = (2)(8) + (-1)(13) = 16 - 13 = 3

따라서 **x = -0.5, y = 3**
