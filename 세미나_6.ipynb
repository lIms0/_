# 6. 개연성 Probability

## 의존 Dependence

: 하나의 사건이 다른 사건에 영향을 받는 관계

- 한 사건의 발생 여부가 다른 사건의 발생 여부에 영향을 줌

- 두 사건 간에 상호 작용이 이으며, 한 사건이 발생할 때 다른 사건의 확률이 변화함

<br>

- ex) 비가 오는 날에 우산을 가지고 있는 사람이 많다.

    - 이 상황에서 비와 우산 사이에 의존 관계가 있음

    - 비가 오는 날에 우산을 가지고 있는 사람의 수가 증가

    - 비가 오지 않는 날에는 우산을 가지고 있는 사람 수가 감소

## 독립 Independence

: 두 사건 간에 상호 작용이 없고, 한 사건의 발생 여부가 다른 사건에 영향을 주지 않는 관계

- 한 사건이 발생하더라도 다른 사건의 확률에는 영향을 주지 않음

<br>
    
- ex) 동전을 두 번 던져서 첫 번째 동전이 앞면일 확률과 두 번째 동전이 앞면일 확률은 서로 독립적이다.
    
    - 첫 번째 동전과 두 번째 동전은 서로 독립적인 사건이므로 첫 번째 동전이 앞면일지 여부가 두 번째 앞면일지 여부에 영햐을 주지 않음

## 조건부 확률 Conditional probability

: 어떤 사건이 일어났다는 조건 하에 해당 사건이 일어날 확률

- 어떤 사건이 일어났을 때 다른 사건이 어떻게 영향을 받는지를 측정하기 위해 사용됨

- P(A|B)와 같이 표현되고, 이는 사건 B가 주어졌을 때 사건 A가 발생할 확률을 나타냄

    - A: 조건부로 발생하는 사건
    - B: 주어진 조건
    
- P(A|B) = P(A∩B) / P(B)

    - P(A∩B): 사건 A와 B가 동시에 발생할 확률을 나타냄
    - P(B): 사건 B가 발생할 확률

<br>
        
- ex) 한 가게에서 노트북을 사는 사람들 중에서 여성인 사람의 비율은 얼마인가?

    - 사건 A: 노트북을 사는 사람이 여성인 경우
    - 사건 B: 노트북을 사는 경우
    - P(A|B): 노트북을 사는 사람들 중에서 여성인 사람의 비율

<br>
    
- ex) 어떤 물리 실험에서 파장 λ의 빛을 내는 레이저를 사용합니다. 그리고 레이저가 정상적으로 작동할 때 레이저 출력이 정상적인 범위 내에 있는 경우를 사건 A로 정의. 또한 실험 중 특정 장치가 고장나서 실험이 제대로 진행되지 않는 상황을 사건 B로 정의

    - P(A|B): 장치가 고장 났을 때 레이저 출력이 정상인 확률
    - P(B): 장치가 고장 날 확률(실험 조건에 따라 다를 수 있음), 0.1이라고 가정
        
        => 장치가 고장날 확률은 0.1
    - P(A∩B): 장치가 고장 났을 때 동시에 레이저 출력이 정상인 확률, 0.02라 가정
    
        => 장치가 고장나고 동시에 레이저 출력이 정상일 확률은 0.02
        
    - 조건부 확률 P(A|B) 계산
    
        => P(A|B) = P(A∩B) / P(B) = 0.02 / 0.1 = 0.2
        
        => 장치가 고장 났을 때 레이저 출력이 정상일 확률: 0.2 = 20%
    
<br>
        
조건부 확률은 사건 간의 상호 작용과 원인과 결과의 관계를 분석하는 데에 매우 유용

주어진 조건에서 어떤 결과가 발생할 가능성을 예측

주어진 결과가 발생했을 때 어떤 조건이 성립할 가능성을 추정



### 베이즈 정리 Baye's theorem

- 조건부 확률을 계산하는 데에 사용되는 중요한 개념

- 사건 간의 관계를 역으로 추론하는 데 활용

- P(A|B) = (P(B|A) * P(A)) / P(B)

    - A, B: 사건 / P(A), P(B): 각각 사건 A와 B가 발생할 확률
    - P(B|A): 사건 A가 주어졌을 때 사건 B가 발생할 확률
    
- 사건 B가 발생한 후 사건 A의 확률을 추정하는 데에 사용

- 초기에는 P(A)와 P(B)를 알고 있어야 하며, P(B|A)는 사전에 알려진 정보를 토대로 추정

-  P(A|B)를 계산하여 사건 B가 발생한 상황에서 사건 A가 발생할 확률을 추론
<br>
- 많은 분야에서 응용되며, 특히 통계, 기계 학습, 인공지능, 의학, 패턴 인식 등에서 활발하게 사용

- 베이지안 통계(Bayesian statistics)의 기초적인 원리로도 활용

## 랜덤 변수 Random Variable

: 결과가 확률적으로 결정되는 변수, 즉 어떤 실험이나 상황에서 여러 가지 값을 가질 수 있는 변수

- 확률 분포를 통해 그 값들의 가능성을 나타냄

- 주로 알파벳 대문자로 표기 ex) X, Y, Z

- 특정 사건에 대한 확률을 나타낼 수 있음

- 각각의 값은 특정 사건이 발생했을 때 해당 사건의 확률을 나타내는 데 사용

- 랜덤 변수는 이산형 discrete 또는 연속형 continuous일 수 있음.

- 이산형 랜덤 변수(discrete random variable)

    - 유한한 또는 셀수 있는 개수의 값 중 하나를 가짐
    
    - ex) 동전 던지기의 결과를 나타내는 랜덤 변수 X는 0 또는 1의 값만을 가지며, 각각은 앞면 또는 뒷면의 결과를 의미
    
- 연속형 랜덤 변수(continuous random variable)

    - 실수 값을 가질수 있음
    
    - ex) 나이를 나타내는 랜덤 변수 Y는 0세부터 무한대까지의 실수 값 중 하나를 가질 수 있음
    
<br>

랜덤 변수는 확률 분포를 통해 그 값들의 가능성을 모델링

이를 통해 우리는 특정 사건이 발생할 확률이나 확률밀도 함수 계산, 예측 및 추론

또한 랜덤 변수를 통해 평균, 분산, **누적 분포 함수** 등과 같은 통계적인 특성 분석

### 누적 분포 함수 Cumulative distribution function, CDF

: 확률 변수의 분포를 나타내는 함수

- 해당 확률 변수가 어떤 특정 값보다 작거나 같을 확률 제공

- 확률 변수 X의 누적 분포함수는 F(x)로 표기되며 다음과 같이 정의

    - F(x) = P(X ≤ x)
    
        - X: 확률 변수
        - x: 특정 값
        - 누적 분포 함수는 x보다 작거나 같은 값의 확률을 나타냄
            
            => x가 확률 변수 X의 가능한 범위 내에 있을 때의 누적 확률을 계산
<br>
- 확률 변수의 분포를 전반적으로 설명하는 중요한 개념

- 확률 변수가 특정 값 이하 또는 이상이 될 확률을 계산할 수 있음

- 누적 분포 함수를 사용하여 평균, 분산 및 기타 통계적 특성을 계산하고 분포의 형태를 시각화할 수도 있음

- 다양한 확률 분포(정규 분포, 이항 분포, **포아송** 분포 등의 확률 분포에서 사용됨)

#### 포아송 분포 Poisson distribution

: 이산형 확률 분포, 일정한 시간 또는 공간 간격에서 발생하는 사건의 수를 모델링하는 데 사용

- 독립적으로 발생하는 사건의 수를 나타내는 데 유용

- 단위 시간 또는 단위 공간에서 사건이 발생하는 비율이 일정한 경우 적용

- 포아송 분포의 확률 질량 함수(probability mass function, PMF)

    - P(X = k) = (e^(-λ) * λ^k) / k!
    
        - X: 포아송 분포를 따르는 확률 변수
        - k: 발생한 사건의 수를 나타내는 값
        - λ: 사건 발생률(rate parameter), 특정 시간 또는 공간 간격에서 사건이 평균적으로 발생하는 횟수
        
- 포아송 분포의 특징

    - 독립성: 각 사건은 독립적으로 발생함
    - 일정한 발생률: 사건 발생률은 일정하며, 시간 또는 공간 간격이 작아질수록 발생하는 사건의 수는 감소
    - 이산형: 발생한 사건의 수는 0부터 무한대까지의 정수로 표현
<br>
포아송 분포는 사건 발생이 독립적이며 발생률이 일정한 많은 현실 세계의 상황을 모델링하는 데 유용

사건 발생 패턴을 이해하고 예측 및 의사 결정을 할 수 있음

## 연속 분포

: 연속적인 값을 가지는 확률 변수를 모델링하는 확률 분포

- 이산 분포와 달리 값들이 이산적이지 않고, 실수 범위 내에서 어떤 값을 가질 수 있음

- 특정 값에 대한 확률을 직접 계산하기보다 구간에 대한 확률을 계산

- 연속 분포를 표현하기 위해 확률 밀도 함수(probability density function, PDF)를 사용

    - **확률 밀도 함수**: 확률 변수가 특정 값 주변에서 발생할 확률
    
        - 적분을 통해 구간에 대한 확률 계산
        - 연속 분포의 확률 밀도 함수는 양수이며, 적분하면 1이 됨
        
<br>

연속 분포는 다양한 형태와 특성을 가질 수 있음

- 정규 분포(normal distribution): 가장 잘 알려진 연속 분포
- 균일 분포(uniform distribution)
- 지수 분포(exponential distribution)
- 감마 분포(gamma distribution)
- 베타 분포(beta distribution)
- t-분포(t-distribution)

<br>

연속 분포의 특성을 이해하고 적절한 분포를 선택하는 것은 데이터 분석과 통계적인 모델링에서 중요함

데이터의 특성을 설명하고 예측할 수 있으며, 확률적인 추론과 의사 결정을 지원

### 정규 분포 normal distribution, 가우시안 분포 (Gaussian distribution)

: **중심 극한의 정리**에 의해 많은 경우 자연스럽게 발생

- 정규 분포는 종 모양의 대칭적인 분포를 가지며, 평균과 표준 편차에 의해 특징 지어짐

- f(x) = (1 / √(2πσ^2)) * e^(-(x - μ)^2 / (2σ^2))

    - x: 확률 변수의 값
    - μ: 평균
    - σ: 표준 편차
    - 확률 변수 x가 특정 값 x 주변에서 발생할 확률
<br>

- 정규 분포의 특징

    - 대칭성: 평균을 중심으로 좌우 대칭적인 형태를 가짐
    - 평균과 중앙값이 일치
    - 표준 편차에 따른 분포 모양 조절: 표준편차가 작을수록 분포는 좁아짐, 표준편차가 클수록 분포는 넓어짐

<br>

- 자연 현상에서 발생하는 무작위 오차, 측정 값의 변동성, 신체 특성의 분포 등을 모델링할 수 있음
- 중심 극한 정리에 따라 많은 독립적인 확률 변수들의 합으로 나타날 수 있다는 특징

#### 중심 극한의 정리 Central limit theorem

: 독립적인 확률 변수들의 합 또는 평균이 정규 분포에 근사하는 현상

- "독립적이고 동일한 분포를 가진 확률 변수들의 합 또는 평균의 표본 크기가 충분히 크다면, 이들의 분포는 정규 분포에 근사하게 됩니다"

- 통계학에서 많은 추론 기법과 가설 검정에 사용됨

- 작은 표본 크기에서부터 추론을 확장하고 모집단에 대한 더 정확한 추정과 예측을 할 수 있음    

### 균일 분포 Uniform distribution

: 주어진 구간에서 모든 값들이 동등한 확률로 발생할 수 있는 분포 > 확률 변수가 특정 구간에서 일정한 확률을 갖는 값을 가질 수 있는 분포

- f(x) = 1 / (b - a)
    
    - a: 구간의 최소값
    - b: 구간의 최대값
    - 위 식은 구간 [a, b]에서 확률 변수 x가 발생할 확률 나타냄

<br>

- 주로 구간 [a, b] 사이에서 발생하는 값들을 모델링하는 데에 사용

    - 구간 내 모든 값들에 대해 동등한 확률을 할당하기 위해 확률 밀도 함수(PDF)가 사용됨

    - 균일 분포에서는 구간 내에서 모든 값들이 동등한 확률을 갖기 때문에 PDF는 해당 구간 내의 길에 반비례하여 일정한 값을 가짐
    
        - 즉, 균일 분포에서 확률 밀도 함수는 구간 [a, b]에서 모든 값에 대해 동일한 값을 할당
        - 구간 외의 값들에 대해서는 확률이 0이므로 PDF는 0을 가지게 됨
        
    - 균일 분포에서는 PDF를 통해 구간 내에서 모든 값들에 대한 동일한 확률을 나타내는 것
    
- 이를 통해 확률 분포를 모델링하고, 구간 내에서 어떤 값이 발생할 확률을 계산

### 지수 분포 Exponential distribution

- 사건이 발생하는 시간을 모델링하는 데 사용

- 주로 사건이 독립적으로 발생하고 발생률(rate)이 일정한 경우에 적용

- 지수 분포의 확률 밀도 함수(PDF)

    - f(x) = λe^(-λx)
    
        - x: 양수인 시간의 간격
        - λ: 발생률 매개 변수(rate parameter), 사건 단위당 사건 발생의 평균 횟수
- 지수 분포의 PDF는 시간 간격이 0보다 크거나 같을 때 확률 밀도를 나타내며 시간 간격이 음수일 경우 확률은 0

- 지수 분포의 특징

    - 이전 사건이 발생한 시간에 상관없이 다음 사건이 발생할 확률은 동일
    - 이전 사건이 발생한 시간에 대한 정보를 기억하지 않고, 다음 사건의 발생 확률은 현재 시간에만 의존
    - 특정 시간 간격이 이미 지난 후에도 다음 사건이 발생할 확률은 처음부터 다시 시작한 것과 동일
    
- 지수 분포는 사건 발생 간격을 모델링하는 데에 유용

### 감마 분포 Gamma distribution

- 양의 실수 값을 가지는 확률 변수의 분포를 모델링하는 데 사용

- 지수 분포와 관련 있음, 지수 분포를 확장하여 다양한 형태의 분포를 모델링할 수 있음

- 감마 분포의 PDF

    - f(x) = (1 / (Γ(k) * θ^k)) * x^(k-1) * e^(-x/θ)
    
        - x는 양의 실수 값
        - k, θ: 감마 분포의 모수
            - k: 모양(shape) 매개 변수, 양의 실수
            - θ: 척도(scale) 매개 변수, 양의 실수
        - Γ(k): 감마 함수(Gamma function), (k-1)!과 유사한 개념
        
- 감마 분포의 특징
    - 감마 분포의 확률 변수는 0 이상의 양의 실수 값으 가짐
    - 모양 매개 변수 k는 분포의 모양을 조절함.
        - k가 작을수록 분포는 비대칭적이고 왼쪽으로 치우져짐
        - k가 커질수록 분포는 대칭적이고 종 모양에 가까워짐
    - 척도 매개 변수 θ는 분포의 척도를 조절
        - θ가 작을수록 분포는 좁아지고 뾰족해짐
        - θ가 커질수록 분포는 넓어지고 평평해짐
        
- 서비스 시간, 수명 시간, 수익의 분포 등을 감마 분포로 모델링

### 베타 분포 Beta distribution

- 0과 1 사이의 값을 가지는 확률 변수의 분포를 모델링하는 데 사용

- 0과 1 사이에서 다양한 형태의 분포를 가질 수 있음

- 베타 분포의 PDF
    - f(x) = (1 / B(α, β)) * x^(α-1) * (1-x)^(β-1)
        - x: 0과 1 사이의 값으로 범위를 가짐
        - α, β: 베타 분포의 모수(parameters), 양의 실수 값, 분포의 모양과 형태를 조절
        -  B(α, β): **베타 함수(beta function)**, α와 β와 관련된 적분
        
- 베타 분포의 특징
    - 베타 분포의 확률 변수는 0과 1 사이의 값을 가짐
    - 모수 α와 β는 분포의 모양과 형태를 조절함.
        - 값을 조절하여 다양한 형태의 분포를 생성할 수 있음.
        - 특히 α = β일 경우 대칭적인 형태의 분포를 가짐
    - 베타 분포의 PDF에는 베타 함수가 포함되어 있으며, 모수 α와 β와 관련된 적분 계산

<br>

-  베이지안 통계, 의사 결정 이론, A/B 테스트 등 다양한 응용 분야에서 널리 활용

    - 베이지안 통계
        - 통계학적으로 관심이 있는 모든 것(모수, 결측치 등)들은 불확실
        - 이 불확실의 정도는 확률로써 표현된다는 가정으로부터 출발
        - 베이지안 통계학의 목표는 주어진 자료(data)를 통하여 관심의 대상이 되는 모수(parameter)의 불확실성(uncertainty)을 확률로써 나타내는 것
    - 의사 결정 이론
        - 불확실한 상황에서 최적의 결정을 하기위한 이론
        - 주로 다수의 가능한 의사결정 대안들 중 하나를 선택
        - 위험 함수의 값에 따라 각 행동마다 효율을 측정하고, 이에 따라 행동의 결정을 가능하게 하는 결정함수를 구하는 이론

### T-분포 T-distribution

- 작은 표본 크기에서 모집단의 평균을 추정하거나 가설 검정을 수행하는 데 사용
- 정규 분포를 근사화하기 위해 개발된 분포로 표본 크기가 작을 때 정확한 추정을 위해 활용

- 표본의 크기(n)에 따라 모양이 변화
    - 표본의 크기가 작을수록 보다 왼쪽으로 치우친 형태를 가짐
- 자유도(degree of freedom)
    - t-분포의 모수
    - 표본의 크기에 따라 결정됨
    - t-분포의 모양과 분산을 제어하는 데 중요한 역할
    
- t-분포 함수의 PDF
    - f(x) = (1 / (sqrt(n) * B(ν/2, 1/2))) * (1 + (x^2/n))^(-(ν+1)/2)
        - x: 확률 변수의 값
        - n: 표본 크기
        - ν: 자유도
        - B(ν/2, 1/2): 베타 함수(beta function)
        
<br>

- t-분포의 특징
    - 표본 크기가 충분히 크다면, t-분포는 정규 분포에 근사
        - 정규 분포를 가정하는 많은 통계 기법에서 t-분포를 사용하여 추정과 가설 검정을 수행
    - t-분포는 표본 크기가 작을수록 보다 왼쪽으로 치우친 형태를 가지게 됨. 표본 크기가 커질수록 t-분포는 정규 분포에 가까워짐
    - t-분포의 자유도는 표본 크기에 의해 결정되며, 분포의 모양과 분산을 제어하는 역할, 자유도가 커질수록 t-분포는 정규 분포에 가까워짐
    
- 표본의 평균을 모집단 평균의 추정값으로 사용할 때, 표본 크기가 작은 경우에는 t-분포를 사용하여 신뢰 구간을 계산
