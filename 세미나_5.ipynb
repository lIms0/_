# 5. 통계 statistics

## 단일 데이터 집합

하나의 변수에 대한 측정값 모음

특정 주제나 현상에 대한 정보 제공, 통계적 분산과 판단에 사용

표의 형태로 표현됨.

각 행은 개별적인 관찰이나 측정값을 나타냄

열은 해당 관찰이나 측정값과 관련된 변수를 나타냄

ex) 학생들의 시험 점수에 대한 데이터 집합

각 행은 개별 학생의 점수를 나타내며, 열은 학생의 이름, 학년, 과목 등과 같은 변수를 포함할 수 있음

<br>

단일 데이터 집합을 사용하여 여러 가지 **통계적 분석**을 수행할 수 있음.

데이터의 중심 경향을 나타내는 측정값 평균, 중앙값, 최빈값을 계산하고, 데이터의 분산과 퍼짐 정도를 나타내는 분산, 표준편차, 범위 등을 계산

시각화를 위해 히스토그램, 상자 그림 등 사용

<br>

**통계적 추론**에도 사용

ex) 표본 데이터를 사용하여 모집단의 특성을 추론하는 작업 가능

모집단에 대한 일반적인 속성이나 특징을 파악하고, 의사결정이나 예측에 도움을 줄 수 있음

<br>

신뢰성과 유의성 평가를 위해 **통계적 가설 검정**을 수행할 수도 있음.

이를 통해 주어진 데이터로부터 얻은 결과가 우연에 의한 것인지, 실제로 통계적으로 유의한지 판단

<br>

-> 결론: 단일 데이터 집합은 하나의 변수에 대한 측정값들의 모음으로, 주제나 현상에 대한 정보를 제공하고 통계적 분석을 위해 사용됨

## 중심 경향 Central Tendency

: 중심에 위치한 대표적인 값 또는 경향을 나타냄.

중심 경향은 데이터 집합의 대표값 제공, 데이터의 특성을 요약하고 해석하는 데 도움을 줌

가장 일반적으로 사용되는 중심 경향의 측정값은 **평균, 중앙값, 최빈값**이다.

### 평균 Mean

- 데이터 집합의 모든 값들을 더한 후, 총 값의 개수로 나누어 계산

- 데이터의 총합을 나타내므로 데이터의 분포를 잘 반영하는 경향

- 극단적인 이상치(outlier) 값이 존재할 경우 평균은 영향을 받음

### 중앙값 Median

- 데이터를 크기 순서로 정렬한 후, 가운데 위치한 값

- 데이터의 크기가 홀수일 경우 중앙값은 정확히 가운데 값, 짝수일 경우 가운데 두 값의 평균

- 데이터의 순서에만 영향을 받으므로 이상치에 덜 민감함

### 최빈값 Mode

- 데이터 집합에서 가장 자주 나타나는 값

- 여러 값들이 동일한 빈도로 나타날 수 있으며, 이 경우 다중 최빈값이 존재할 수 있음

- 최빈값은 주로 범주형 데이터에서 사용됨

<br>

중심 경향의 선택은 데이터의 특성과 목적에 따라 달라짐

단독으로만 사용하는 것이 아니라 데이터의 특성과 다른 통계적 측정값과 함께 고려하는 것이 중요함

### 분산 Variance

: 데이터의 흩어진 정도나 변동성을 나타내는 지표

### 표준편차 Standard Deviation

: 분산의 양의 제곱근

분산과 마찬가지로 데이터의 변동성을 측정하는 데 사용

<br>

-> 높은 분산 또는 표준편차는 데이터가 평균 주의에 퍼져 있는 것을 나타내며, 낮은 값은 데이터가 평균에 가까이 모여 있는 것을 의미

### 표본과 모집단 Sample and Population

: 통계 분석에서는 일반적으로 표본과 모집단의 개념을 사용

<br>

- 표본

: 모집단에서 추출된 작은 데이터 집합

통계적 추론을 수행하기 위해 사용됨

- 모집단

: 연구 대상이 되는 전체 집단을 나타냄

통계적 분석 결과를 일반화하고 모집단에 대한 추론을 가능하게 함

### 표본 크기와 통계적 신뢰도 Sample Size and Statistical Power

충분한 표본 크기를 확보하면 통계 분석 결과의 신뢰도가 증가하고 통계적 검정의 감지력(power)이 향상됨

통계적 감지력은 유의한 효과나 차이를 감지할 수 있는 능력을 나타냄

실험 또는 연구 결과의 신뢰성을 높이기 위해 고려되어야 함

## 통계적 가설 검정 Statistical Hypothesis Testing

: 주어진 데이터를 기반으로 특정 가설을 검증하는 통계적 절차

이 절차를 통해 주장하고자 하는 가설에 대한 증거를 평가하고, 통계적으로 유의한 차이가 있는지를 판단

가설검정은 일반적으로 귀무가설(null hypothesis)과 대립가설(alternative hypothesis)을 설정하고, 유의수준(significance level)과 통계적 검정 기법을 사용하여 가설을 평가함

## 상관관계 

- 두 변수 간의 관련성 또는 연관성을 측정하는 통계적 개념

- 일반적으로 상관계수(correlation coefficient)를 사용하여 상관관계를 나타냄

- 상관계수는 -1에서 1까지 범위를 가지며, 값의 부호와 크기에 따라 상관관계의 강도와 방향성을 나타냄

- 상관계수가 양수일 경우 양의 상관관계(두 변수가 함께 증가 또는 감소)

- 상관계수가 음수일 경우 음의 상관관계(한 변수가 증가하면 다른 변수는 감소)

- 0에 가까울수록 상관관계가 약해짐

- 인과관계를 나타내지 않으며, 두 변수 간의 관련성만을 나타내는 것

## 인과관계 Causation

- 한 변수가 다른 변수에 직접적인 영향을 미치는 관계

- 확인하기 위해서는 실증적인 연구 디자인, 실험 또는 복잡한 분석 방법이 필요

- 이를 통해 통계적으로 유의한 인과관계를 확인할 수 있음

- 인과관계를 파악하기 위해 다음 요소를 고려해야 함

    1. 시간적 선행성
    
    : 인과관계에서 원인은 결과보다 먼저 발생해야 함
    
    원인 변수의 변화가 결과 변수의 변화를 앞서야 함.
    
    이는 원인과 결과의 시간적 순서를 파악하는 것을 의미
    
    ex) 햇빛이 쨍쨍하다고 해서 풀이 자라는 것이 아니지만 풀이 자라면서 햇빛을 받는 것은 가능
    
    
    2. 인과적 메커니즘(기계적 구조)
    
    : 원인과 결과 간의 명확한 메커니즘이나 작용 방식이 존재해야 함
    
    원인과 결과 간에는 원인이 결과를 직접적으로 설명하고 영향을 미치는 일련의 과정이 있어야 함
    
    
    3. 대체 설명의 제거
    
    : 원인과 결과 사이에 다른 요인이 존재하지 않거나 그 영향을 제거해야 함.
    
    다른 요인들이 원인과 결과 사이의 관계를 설명할 수 있는 대체 설명이 될 수 있기 때문
    
    이를 위해 대체 설명을 제거하기 위한 실험 설계나 통계적 방법을 사용하여 원인 변수의 영향을 독립적으로 분석하고 평가할 필요가 있음
    
    
    
-> 통계적인 상관관계만으로는 결론을 내릴 수 없는 더 높은 수준의 연구와 분석을 필요로 함

통계적 방법을 통해 원인과 결과 간의 인과 관계를 분석할 때에도 변수의 조작과 제어, 대체 설명의 제거 등에 주의를 기울여야 함

두 변수 간의 관계를 이해하고 예측하기 위해 중요함. 인과관계를 확인하기 위해서는 신중한 연구 디자인과 적절한 분석 방법을 사용하여 다른 가능성을 제거하고 결과를 타당하게 해석해야 함.

상관관계와 인과관계는 서로 다른 개념

상관관계가 있다고 해서 반드시 인과관계가 있는 것 xxx

인과관계를 확인하기 위해서는 심층적인 연구가 필요

데이터 분석에서는 상관관계와 인과관계를 구분하여 해석하는 것이 중요

## 이상값 Outlier

- 데이터 집합에서 다른 값들과 동떨어진 극단적인 값을 의미

- 이상값은 데이터의 분포를 왜곡시킬 수 있으며, 통계적 분석 결과에 영향을 미칠 수 ㅇ있음

- 다양한 원인에 의해 발생할 수 있음 ex) 실험 오류, 잘못된 데이터 입력, 자연적인 변동 등

- 데이터 집합의 분포를 시각화하거나 통계적으로 식별하는 방법으로 찾을 수 이음

- 처리하는 방법
    ) 제거, 교정, 대체
    데이터 분석의 목적과 특성에 따라 결정됨
    
<br>

적절한 상관관계 분석은 두 변수 간의 관계를 파악하고 예측에 유용한 정보를 제공할 수 있음.

이상값의 처리는 데이터의 정확성과 신뢰성을 높이고, 분석 결과의 왜곡을 방지하기 위해 필요함.

## 심슨의 역설 Simpson's Paradox

: 데이터의 그룹이 서로 다른 방식으로 분할되었을 때, 서로 상반된 결과를 보여 줄 수 있는 통계적 현상

- 이러한 역설은 때때로 데이터의 집단 간의 상관관계를 왜곡시킬 수 있으며, 잘못된 결론을 도출할 수 있음

<br>

주로 다음과 같은 상황에서 나타남

1. 데이터가 서로 다른 하위 그룹으로 분할되어 있고, 각 하위 그룹의 크기나 구성이 상이한 경우

2. 각 하위 그룹 내에서의 상관관계가 전체 데이터와는 다른 방향이나 강도를 가지는 경우

<br>

주로 통계적 분석에서 관찰되는 현상이지만 실제 세계에서도 다양한 예시를 찾아볼 수 있음.

ex) 한 연구에서 전체 집단에서 A 그룹이 B 그룹보다 더 높은 비율의 성공을 보일 수 있음. 그러나 A 그룹과 B 그룹을 세부적으로 나누어보면, 각 하위 그룹에서는 B 그룹이 더 높은 비율의 성공을 보일 수도 있음. 이 경우 전체 데이터에서는 A 그룹이 더 우수한 것처럼 보이지만, 하위 그룹 수준에서는 B 그룹이 더 우수한 결과를 보이는 모순이 발생.

<br>

- 데이터를 분석할 때 주의해야 할 중요한 경고 신호임

그룹 간의 상관관계를 평가할 때는 추가적인 변수나 하위 그룹 간의 차이점을 고려해야 함

하위 그룹간의 크기 차이나 다른 요인들이 결과에 영향을 미칠 수 있으며, 이를 무시하면 잘못된 결론을 도출할 수 있음

<br>

-> 통계적 분석에서 데이터의 해석을 조심하고 항상 전체적인 상황을 고려해야 함을 상기시키는 개념.

   데이터 분석에서는 전체적인 맥락과 다양한 변수 간의 관계를 고려하여 결과를 해석하는 것이 필요
